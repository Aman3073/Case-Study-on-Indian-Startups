# Question-1

```python
#importing required modules
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt

funding = pd.read_csv("C:\\Users\\hp\\Data-science-and-machine-learning\\Workingwithfile//startup_funding.csv")
df = funding.copy()
df.dropna(subset = ["CityLocation"],inplace = True)
df["CityLocation"] = df["CityLocation"].replace("bangalore","Bangalore")
df["CityLocation"] = df["CityLocation"].replace("Delhi","New Delhi")


def valid_city(city):
    city = str(city)
    if city != '':
        c_n = city.split('/')[0].strip()
        if c_n == "New Delhi" or c_n == "Gurgaon" or c_n == "Noida" or c_n == "Mumbai" or c_n == "Bangalore":
            return c_n

df["CityLocation"] = df["CityLocation"].apply(valid_city)
df["CityLocation"] = df["CityLocation"].replace(["Gurgaon","Noida","New Delhi","Delhi"],"NCR")
final_output = df["CityLocation"].value_counts()
city_name = final_output.index
city_count = final_output.values
for i in range(3):
    print(city_name[i],city_count[i])
plt.bar(city_name,city_count,width = 0.5)
plt.title("Locations to set the Startup",color = "green")
plt.xlabel("Top City",color = "blue")
plt.ylabel("Total no. of investment",color = "blue")
plt.show()
```

# Question-2

```python
#importing required modules
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt

startup = pd.read_csv("C:\\Users\\hp\\Data-science-and-machine-learning\\Workingwithfile//startup_funding.csv")
df = startup.copy()
df.dropna(subset = ["InvestorsName"],inplace = True)
new =df["InvestorsName"].values

d = {}
for ele in new:
    final = ele.split(",")
    for i in final:
        key = i.strip()
        if key in d:
            d[key] +=1
        else:
            d[key] = 1
a = sorted(d.items(), key = lambda kv:(kv[1], kv[0]),reverse = True)[:5]
invester_list = []
total_investmen = []
for ele in a:
    invester_list.append(ele[0])
    total_investmen.append(ele[1])
    print(ele[0],ele[1])



plt.bar(invester_list,total_investmen,width = 0.5)
plt.xticks(rotation = 40)
plt.title("Top 5 Investors")
plt.xlabel("Top Investor Name",color = "blue")
plt.ylabel("Total no. of investment",color = "blue")
plt.show()
```

# Question-3

```python
#importing required modules
import pandas as pd
import csv
import matplotlib.pyplot as plt
startup = pd.read_csv("C:\\Users\\hp\\Data-science-and-machine-learning\\Workingwithfile//startup_funding.csv")
df = startup.copy()


df.dropna(subset = ["InvestmentType"],inplace = True)
df.dropna(subset = ["InvestorsName"],inplace = True)

df["StartupName"] = df["StartupName"].replace(["Ola Cabs","Olacabs"],"Ola")
df["StartupName"] = df["StartupName"].replace("Flipkart.com","Flipkart")
df["StartupName"] = df["StartupName"].replace(["Oyo Rooms","OYOfit","OyoRooms","Oyorooms","OYO Rooms"],"Oyo")
df["StartupName"] = df["StartupName"].replace(["Paytm Marketplace"],"Paytm")


d = {}
for index,row in df.iterrows():
    s = row['InvestorsName']
    v = row['StartupName']
    
    for ele in s.split(","):
        value = v
        key = ele.strip()
    
        if key in d:
            if value not in d[key]:
                d[key].append(value)
        else:
            d[key] = [value]
            
new = {}
for ele in d:
    if ele != "":
        new[ele] = len(d[ele])
a = sorted(new.items(), key = lambda kv:(kv[1], kv[0]),reverse = True)[:5]
invester_list = []
total_investmen = []
for ele in a:
    invester_list.append(ele[0])
    total_investmen.append(ele[1])
    print(ele[0],ele[1])



plt.bar(invester_list,total_investmen,width = 0.5)
plt.xticks(rotation = 40)
plt.xlabel("Top Investor Name",color = "blue")
plt.ylabel("Total no. of investment",color = "blue")
plt.title("Top 5 Investors in Distinct companies")
plt.show()
```

```python
#importing required modules
import pandas as pd
import csv
import matplotlib.pyplot as plt
startup = pd.read_csv("C:\\Users\\hp\\Data-science-and-machine-learning\\Workingwithfile//startup_funding.csv")
df = startup.copy()


df.dropna(subset = ["InvestmentType"],inplace = True)
df.dropna(subset = ["InvestorsName"],inplace = True)

df["InvestmentType"] = df["InvestmentType"].replace("PrivateEquity","Private Equity")
df["InvestmentType"] = df["InvestmentType"].replace("SeedFunding","Seed Funding")
df["InvestmentType"] = df["InvestmentType"].replace("Crowd funding","Crowd Funding")

df["StartupName"] = df["StartupName"].replace(["Ola Cabs","Olacabs"],"Ola")
df["StartupName"] = df["StartupName"].replace("Flipkart.com","Flipkart")
df["StartupName"] = df["StartupName"].replace(["Oyo Rooms","OYOfit","OyoRooms","Oyorooms","OYO Rooms"],"Oyo")
df["StartupName"] = df["StartupName"].replace(["Paytm Marketplace"],"Paytm")

df = df[(df["InvestmentType"] ==  "Seed Funding") | (df["InvestmentType"] ==  "Crowd Funding")]
d = {}
for index,row in df.iterrows():
    s = row['InvestorsName']
    v = row['StartupName']
    
    for ele in s.split(","):
        value = v
        key = ele.strip()
    
        if key in d:
            if value not in d[key]:
                d[key].append(value)
        else:
            d[key] = [value]
    
new = {}
for ele in d: 
    if (ele != "Undisclosed Investors") and (ele != "") and (ele != "Undisclosed investors"):
        new[ele] = len(d[ele])
a = sorted(new.items(), key = lambda kv:(kv[1], kv[0]),reverse = True)[:5]
invester_list = []
total_investmen = []
for ele in a:
    invester_list.append(ele[0])
    total_investmen.append(ele[1])
    print(ele[0],ele[1])



plt.bar(invester_list,total_investmen,width = 0.5)
plt.xticks(rotation = 40)
plt.xlabel("Top Investor Name",color = "blue")
plt.ylabel("Total no. of investment",color = "blue")
plt.title("Top 5 Investors in Distinct companies  with Investment Type as Crowd Funding or Seed Funding")
plt.show()
```
